```python
h = 6.62607004e-034 # m2 kg / s
c = 299792458 # m / s
G = 6.67408e-11 # m3 kg-1 s-2
l = (h*G/c**3)**0.5
t = (h*G/c**5)**0.5
m = (c*h/G)**0.5
print(f"Foundamental length: {l}[m]")
print(f"Foundamental time: {t}[s]")
print(f"Foundamental mass: {m}[kg]")
print("************************************************")
print(f"Foundamental speed: {l/t}[m/s]")
print(f"Foundamental acceleration: {l/t**2}[m/s2]")
print(f"Foundamental Force: {m*l/t**2}[N]")
print(f"Foundamental Energy: {m*(l/t)**2}[J]")
print(f"Foundamental Power: {(m*(l/t)**2)/t}[W]")
print(f"Foundamental pressure: {(m*l/t**2)/l**2}[N/m2]")
```
